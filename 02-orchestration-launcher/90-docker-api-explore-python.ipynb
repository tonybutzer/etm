```python
def _return_last_write(container):
    bytesThing = container.logs()
    newStringThing = bytesThing.decode(encoding='UTF-8')
    lines = newStringThing.split('\n')
    last_write='bogus'
    for l in lines:
        if 'Writing' in l:
            #print(l)
            last_write = l
    return (last_write)
```

```python
for c in running_containers:
    lw = _return_last_write(c)
    print(lw)
```

```python
lws=[]
for c in running_containers:
    lw = _return_last_write(c)
    #print(lw)
    lws.append(lw)
```
