```python
import pandas as pd  
import os

with open('MIROC5_historical_future8.5_cloudVegET.sh') as fo:
    lines = fo.readlines()

# Extract name of netCDF from shell script
outs=[]
for line in lines:
    if '-O' in line:
        out=line.split('-O')[1][1:]
        out = out.rstrip()
        out = out.split('.nc')[0]
        print(out)
        outs.append(out)

# break down the name of each file into each metadata component
dfis=[]
for out in outs:
    a=out.split('_')
    dfi = {'dataset':a[0], 'measure':a[1], 'model':a[2], 'r6':a[3], 'epoch':a[4], 'start_year':a[5], 'end_year':a[6],
          'region':a[7], 'frequency':a[8]}
    dfis.append(dfi)
```
