```python
import pandas as pd 
```

```python
def _calc_pct(out, start_year, end_year, product):
    
    mydict = st_build_year_counts(out, start_year, end_year, product)

    for key in sorted(mydict):
        print( "%s: %s" % (key, mydict[key]))
    new = pd.DataFrame.from_dict(mydict, orient='index',columns=['scene_count'])
    mymean = new["scene_count"].mean()
    ideal = 377

    pct = mymean/ideal

    pct = pct * 100

    answer = str(round(pct,2))
    print(answer, " PERCENT Complete for ", product)
    print("PCT",pct)
    return pct
```

```python
out='enduser/DelawareRiverBasin/drb150/'
start_year = 1950
end_year = 2099
products = ['etc', 'etasw', 'dd', 'srf', 'netet']
pct_hash = {}
for product in products:
    pct = _calc_pct(out, start_year, end_year, product)
    pct_hash[product] = pct
```

```python
pctdf = pd.DataFrame.from_dict(pct_hash, orient='index',columns=['PCT'])
```
